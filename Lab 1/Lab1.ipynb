## 1. Data Set Explanation and Preview

The **Wine Quality (Red)** dataset comes from the UCI Machine Learning Repository. It contains **1,599** samples of red wine from the Vinho Verde region in Portugal. Each sample was evaluated by experts on a **quality score** from 0 (very poor) to 10 (excellent). 

**Purpose:**  
Use simple chemical measurements to predict how good a wine is.

**Key Features:**  
- **Fixed acidity**  
- **Volatile acidity**  
- **Citric acid**  
- **Residual sugar**  
- **Chlorides**  
- **Free sulfur dioxide**  
- **Total sulfur dioxide**  
- **Density**  
- **pH**  
- **Sulphates**  
- **Alcohol**  
- **Quality** (target variable)


```python
import pandas as pd

# Load the CSV
df = pd.read_csv('winequality-red.csv', sep=';')

# Preview
df.head()
```

---

## 2. Bar Graph Visualization: Quality Counts

This bar graph shows how many wines received each quality score.


```python
# bar graph of quality counts
import matplotlib.pyplot as plt

quality_counts = df['quality'].value_counts().sort_index()

plt.figure(figsize=(8,6))
quality_counts.plot(kind='bar')
plt.title('Wine Quality Counts', fontsize=16)
plt.xlabel('Quality Score', fontsize=14)
plt.ylabel('Number of Samples', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('VandanAmin_WineQualityCounts.png')
plt.show()
```

**Observation:**  
Most wines are rated around 5â6, with fewer at the extremes (3 or 8+).


---

## 3. Line Graph Visualization: Avg. Alcohol by Quality


```python
# line graph of average alcohol content per quality
avg_alc = df.groupby('quality')['alcohol'].mean()

plt.figure(figsize=(8,6))
avg_alc.plot(marker='o', linestyle='-')
plt.title('Average Alcohol Content by Quality', fontsize=16)
plt.xlabel('Quality Score', fontsize=14)
plt.ylabel('Avg. Alcohol (% by vol)', fontsize=14)
plt.grid(True)

plt.savefig('VandanAmin_AvgAlcoholByQuality.png')
plt.show()
```

**Observation:**  
Higherâquality wines tend to have slightly higher alcohol content on average.


---

## 4. Pie Chart Visualization: Alcohol Level Categories

Here we categorize alcohol into **Low** (â¤10%), **Medium** (10â12%), **High** (>12%) and show their proportions.


```python
# pie chart of alcohol categories
bins = [0, 10, 12, df['alcohol'].max()]
labels = ['Low (â¤10%)', 'Medium (10â12%)', 'High (>12%)']
df['alc_category'] = pd.cut(df['alcohol'], bins=bins, labels=labels, include_lowest=True)

cat_counts = df['alc_category'].value_counts().reindex(labels)

plt.figure(figsize=(6,6))
cat_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Alcohol Level Categories', fontsize=16)
plt.ylabel('')

plt.savefig('VandanAmin_AlcoholCategories.png')
plt.show()
```

**Observation:**  
The majority of samples fall into the Low (â¤ 10%) alcohol category.


---

## 5. Box Chart Visualization: pH by Quality

A box plot to compare the pH distribution across different quality scores.


```python
# box plot of pH by quality
plt.figure(figsize=(8,6))
df.boxplot(column='pH', by='quality')
plt.title('pH Distribution by Quality', fontsize=16)
plt.suptitle('')  # remove the automatic âBoxplot grouped by qualityâ subtitle
plt.xlabel('Quality Score', fontsize=14)
plt.ylabel('pH', fontsize=14)

plt.savefig('VandanAmin_pHBoxByQuality.png')
plt.show()
```

**Observation:**  
Higherâquality wines show a slightly narrower pH range compared to lowerâquality ones.
